Commercial and industrial algae cultivation has numerous uses, including ... fertilizers, bioplastics, chemical feedstock (raw material)[s1]...

Cultured microalgae already contribute to a wide range of sectors in the emerging [bioeconomy](https://en.wikipedia.org/wiki/Bioeconomy).[s2]

For centuries seaweed has been used as fertilizer. It is also an excellent source of potassium for manufacture of potash and potassium nitrate. Some types of microalgae can be used this way as well.[s3]

Cultivated algae serve many other purposes, including cosmetics,[s57][further explanation needed] animal feed,[s57][further explanation needed] bioplastic production, dyes and colorant production, chemical feedstock production, and pharmaceutical ingredients.[further explanation needed]

To grow it:

- Macroalgae, ie kelp, are farmed: can be harvested from natural growth or farmed in shallowly-submerged fields. Microalgae can be grown in varieties of (phyto)bioreactors, or in various open or closed lagoons and "raceway ponds". Modernly, High-Rate Algal Ponds (HRAPs)[s4] are being used. Most methods apply to micro biofilm-forming organisms and to macroalgae. Our method might be a closed raceway pond, where seawater filtered through a mesh against cells and macromolecules spills over microalgae in trays hoisted on land or near the sea but not necessarily submerged in the ocean. 
- There are various methods for harvesting, including chemicals, "microscreens, by centrifugation, by flocculation and by froth flotation." ie, there's been plenty of work here to buld off of.

other notes:
- Another approach [to grow it] is to use a special medium which excludes other organisms, including invasive algae. For example, [Dunaliella](https://en.wikipedia.org/wiki/Dunaliella) is a commonly grown genus of microalgae which flourishes in extremely salty water that few other organisms can tolerate. 
- Some living algae and bacteria also produce odorous chemicals, particularly certain cyanobacteria (previously classed as blue-green algae) such as [Anabaena](https://en.wikipedia.org/wiki/Anabaena) 
- The seven most cultivated taxa are Eucheuma spp., Kappaphycus alvarezii, Gracilaria spp., Saccharina japonica, Undaria pinnatifida, Pyropia spp., and Sargassum fusiforme. Eucheuma and K. alvarezii are attractive for carrageenan (a gelling agent); Gracilaria is farmed for agar; the rest are eaten after limited processing. Seaweeds are different from mangroves and seagrasses, as they are photosynthetic algal organisms and are non-flowering. [1](https://en.wikipedia.org/wiki/Eucheuma), [6](https://en.wikipedia.org/wiki/Gracilaria), [2](https://en.wikipedia.org/wiki/Kappaphycus_alvarezii), [3](https://en.wikipedia.org/wiki/Saccharina_japonica), [4](https://en.wikipedia.org/wiki/Undaria_pinnatifida), [5](https://en.wikipedia.org/wiki/Pyropia),  [7](https://en.wikipedia.org/wiki/Sargassum_fusiforme), 
- Open pond systems are cheaper to construct, at the minimum requiring only a trench or pond. Large ponds have the largest production capacities relative to other systems of comparable cost. Also, open pond cultivation can exploit unusual conditions that suit only specific algae. For instance, [Dunaliella salina](https://en.wikipedia.org/wiki/Dunaliella_salina) grow in extremely salty water; these unusual media exclude other types of organisms, allowing the growth of pure cultures in open ponds. 
- @ V-shaped photobioreactors: V-shaped plastic bags are effective for growing a variety of algae species, including Chlorella, Spirulina, and Nannochloropsis. The growth rate and biomass yield of Chlorella vulgaris in V-shaped plastic bags was found to be higher than any other shaped plastic bags. [s6]

Other uses

Microalgae are a source of valuable molecules such as isotopes i.e. chemical variants of an element that contain different neutrons. Microalgae can effectively incorporate isotopes of carbon (13C), nitrogen (15N) and hydrogen (2H) into their biomass.[41] 13C and 15N are used to track the flow of carbon between different trophic levels/food webs.[42] Carbon, nitrogen and sulphur isotopes can also be used to determine disturbances to bottom dwelling communities that are otherwise difficult to study.[42]

Microalgae are unicellular aquatic microorganisms that can grow as single cells, chains, or tiny colonies. They are found in both freshwater and marine environments which can also perform photosynthesis, hence referred to as photoautotrophs A recent analysis projects that the worldwide market for algal products would reach a value of more than 5.38 billion US dollars by 2025. Market value for microalgal fuels is anticipated to rise significantly. One underutilized resource for high-value products is microalgae. Microalgae might emerge as a promising and sustainable food, feed, and nutraceutical resource to satisfy current market demands due to the growing desire for natural products, customized functional foods, and increased consumer awareness about health advantages. They are abundant sources of several substances that can strengthen diets rich in nutrients, boost immunity, and improve overall health. Moreover, microalgae are a rich source of lipids, which makes them attractive candidates in the global hunt for energy substitutes. **However, a significant disadvantage of the state of microalgal research is the inability to transfer in vitro results to an industrial scale. Therefore, finding a resilient microalgal species that can generate a manageable amount of high-value chemicals is crucial.** This chapter covers the fundamentals of microalgae technology, outlining its key features and providing a thorough presentation of microalgae commercially significant compounds and their potential usage.
